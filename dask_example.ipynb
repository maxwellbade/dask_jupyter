```python
import numpy as np
import pandas as pd
import dask.dataframe as dd

import dask.array as da
from dask.distributed import Client

client = Client()
client
```

```python
x = da.random.random((10000, 10000), chunks=(1000, 1000))

print(x)

y = x + x.T
z = y[::2, 5000:].mean(axis=1)

print(z)

z.compute()

y = y.persist()

%time y[0, 0].compute()

%time y.sum().compute()

x = da.random.random((10000, 10000), chunks="16 MiB") 

print(x)

y = (x+x.T) - x.mean(axis=0)

print(y.sum().compute())

x = da.random.random((10000, 10000, 10), chunks=(1000, 1000, 5))
y = da.random.random((10000, 10000, 10), chunks=(1000, 1000, 5))
z = (da.arcsin(x) + da.arccos(y)).sum(axis=(1, 2))
z.persist()

%time
x = da.random.random(10_000_000, chunks=(10_000_000,))
print(x.sum().compute())

df = dd.read_csv('test.csv')
print(df.head())

%time
print(df.prices.sum().compute())

x = da.random.random((10000, 10000), chunks=(1000, 1000))

print(x)

y = x + x.T
z = y[::2, 5000:].mean(axis=1)

print(z)

z.compute()

y = y.persist()

%time y[0, 0].compute()

%time y.sum().compute()

x = da.random.random((10000, 10000), chunks="16 MiB") 

print(x)

y = (x+x.T) - x.mean(axis=0)

print(y.sum().compute())

x = da.random.random((10000, 10000, 10), chunks=(1000, 1000, 5))
y = da.random.random((10000, 10000, 10), chunks=(1000, 1000, 5))
z = (da.arcsin(x) + da.arccos(y)).sum(axis=(1, 2))
z.persist()

%time
x = da.random.random(10_000_000, chunks=(10_000_000,))
print(x.sum().compute())

df = dd.read_csv('test.csv')
print(df.head())

%time
print(df.prices.sum().compute())

x = da.random.random((10000, 10000), chunks=(1000, 1000))

print(x)

y = x + x.T
z = y[::2, 5000:].mean(axis=1)

print(z)

z.compute()

y = y.persist()

%time y[0, 0].compute()

%time y.sum().compute()

x = da.random.random((10000, 10000), chunks="16 MiB") 

print(x)

y = (x+x.T) - x.mean(axis=0)

print(y.sum().compute())

x = da.random.random((10000, 10000, 10), chunks=(1000, 1000, 5))
y = da.random.random((10000, 10000, 10), chunks=(1000, 1000, 5))
z = (da.arcsin(x) + da.arccos(y)).sum(axis=(1, 2))
z.persist()

%time
x = da.random.random(10_000_000, chunks=(10_000_000,))
print(x.sum().compute())

df = dd.read_csv('test.csv')
print(df.head())

%time
print(df.prices.sum().compute())

x = da.random.random((10000, 10000), chunks=(1000, 1000))

print(x)

y = x + x.T
z = y[::2, 5000:].mean(axis=1)

print(z)

z.compute()

y = y.persist()

%time y[0, 0].compute()

%time y.sum().compute()

x = da.random.random((10000, 10000), chunks="16 MiB") 

print(x)

y = (x+x.T) - x.mean(axis=0)

print(y.sum().compute())

x = da.random.random((10000, 10000, 10), chunks=(1000, 1000, 5))
y = da.random.random((10000, 10000, 10), chunks=(1000, 1000, 5))
z = (da.arcsin(x) + da.arccos(y)).sum(axis=(1, 2))
z.persist()

%time
x = da.random.random(10_000_000, chunks=(10_000_000,))
print(x.sum().compute())

df = dd.read_csv('test.csv')
print(df.head())

%time
print(df.prices.sum().compute())

x = da.random.random((10000, 10000), chunks=(1000, 1000))

print(x)

y = x + x.T
z = y[::2, 5000:].mean(axis=1)

print(z)

z.compute()

y = y.persist()

%time y[0, 0].compute()

%time y.sum().compute()

x = da.random.random((10000, 10000), chunks="16 MiB") 

print(x)

y = (x+x.T) - x.mean(axis=0)

print(y.sum().compute())

x = da.random.random((10000, 10000, 10), chunks=(1000, 1000, 5))
y = da.random.random((10000, 10000, 10), chunks=(1000, 1000, 5))
z = (da.arcsin(x) + da.arccos(y)).sum(axis=(1, 2))
z.persist()

%time
x = da.random.random(10_000_000, chunks=(10_000_000,))
print(x.sum().compute())

df = dd.read_csv('test.csv')
print(df.head())

%time
print(df.prices.sum().compute())
```
